## Matplotlib Assignment

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig, axes = plt.subplots( )

axes.set_yticks([0, 50, 100, 150,200])

axes.plot(x, y, 'b')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

```python
fig, ax = plt.subplots()

ax1 = fig.add_axes( [0,0,1,1])
ax2 = fig.add_axes( [0.2,0.5,.2,.2])
```

```python
ax1.plot(x, y, 'r')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_yticks([0, 50, 100, 150, 200 ])


ax2.plot(y, x, 'r')
ax2.set_xlabel('x')
ax2.set_ylabel('y');
```

```python
fig, axes = plt.subplots()

axes1 = fig.add_axes( [0,0,1,1])
axes2 = fig.add_axes( [0.2,0.5,.4,.4])
```

```python
axes1.plot(x,z,'b')
axes1.set_xlabel('X')
axes1.set_ylabel('Z')

axes2.plot(x,y,'b')
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('zoom')
axes2.set_xlim([20,22])
axes2.set_ylim([30,50])
```
